```python
import pandas as pd
# how to retrieve certain cols
housing =pd.read_csv("https://raw.githubusercontent.com/ProfHan/ITEC4700/refs/heads/main/Housing.csv", usecols=['longitude', 'latitude', 'median_house_value'] )
# you can use seaborn for visualization
# what else can you use to present the third variable?
import seaborn as sns
sns.scatterplot( data=housing, x='longitude', y='latitude', hue='median_house_value'     )
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(housing[['latitude', 'longitude']], housing['median_house_value'], test_size =0.2, random_state=0)

```

```python
from sklearn import preprocessing
x_train_norm = preprocessing.normalize(x_train)
x_test_norm = preprocessing.normalize(x_test)
```

```python
from sklearn.cluster import KMeans

Kmeans = KMeans (n_clusters =3, random_state=0, n_init= "auto")
Kmeans.fit (x_train_norm)
# K-Means starts from random initial cluster centers.
#To get a good solution, it tries multiple random initializations and keeps the best one.

#n_init controls how many times to restart the algorithm.
```

```python
sns.scatterplot (data=x_train, x="latitude", y="longitude", hue=Kmeans.labels_)
```
